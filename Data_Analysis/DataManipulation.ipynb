## Data Manipulation and Analysis with Pandas


```python
import pandas as pd
```

```python
df=pd.read_csv('employment.csv')
##fetch the first five rows
df.head(5)
```

```python
df.describe()
```

```python
df.dtypes
```

```python
##Handling missing values
df.isnull().any() ##anyt(axis =1)
```

```python
df.isnull().sum() ##tells where how many missing values are there
```

```python
df_filled=df.fillna(5) ##wherever there is a missing value it will be replaced by 5
```

```python
##fill missing values with the mean of the column
df['Data_value_new']=df['Data_value'].fillna(df['Data_value'].mean())
df
```

```python
df.dtypes
```

```python
##Renaming columns
df.rename(columns={'STATUS':'SATAYAUA'})
```

```python
##change data types
df['Value_new'] = df['Magnitude'].astype(float)
df.head()
```

```python
df['New Value']=df['Period'].apply(lambda x:x*2)
df.head()
```

```python
## Data Aggregating and Grouping

df.head()
```

```python
grouped_mean=df.groupby('Group')['Data_value'].mean()
print(grouped_mean)
```

```python
group_sum=df.groupby(['Group','Series_title_2'])['Data_value'].sum()
print(group_sum)
```

```python
df.groupby(['Group','Series_title_2'])['Data_value'].mean()
```

```python
##aggregate multiple functions


df1=pd.read_csv('organizations-100.csv')
##fetch the first five rows
df1.head(5)

grouped_agg=df1.groupby('Name')['Number of employees'].agg(['mean','sum','count'])
grouped_agg
```

```python
##Merging and joining dataframes
#create sample Dataframes
df2=pd.DataFrame({'Key': ['A','B','C'],'Value1': [1,2,3]})
df3=pd.DataFrame({'Key': ['A','B','D'],'Value2': [4,5,6]})
```

```python
df2
```

```python
df3
```

```python
##Merge Dataframe and the key columns
pd.merge(df2,df3,on="Key", how ="inner") ##On --> which column name; 
```

```python
pd.merge(df2,df3,on="Key", how ="outer")
```

```python
pd.merge(df2,df3,on="Key", how ="left") ##df2 is given priority here
```

```python
pd.merge(df2,df3,on="Key", how ="right")##df3 is given priority here
```
